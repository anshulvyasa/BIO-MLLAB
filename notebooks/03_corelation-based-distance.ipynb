```python
# getting back our pre-processed data
import pandas as pd
import numpy as np
from pathlib import Path

DATA_PROCESSED = Path("../data/processed")
gse_id = "GSE2034"

expr = pd.read_parquet(DATA_PROCESSED / f"{gse_id}_preprocessed.parquet")
print("Matrix shape:", expr.shape)
```

```python
# correlation between rows â†’ so use .corr on transpose
corr_matrix = expr.T.corr()
print("Correlation matrix shape:", corr_matrix.shape)
```

```python
dist_matrix = 1 - corr_matrix
```

```python
dist_matrix.to_parquet(DATA_PROCESSED / f"{gse_id}_distance_matrix.parquet")
```
